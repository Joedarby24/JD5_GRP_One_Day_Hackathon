# Time for Dummies, EDA, & Feature Engineering

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
```

```python
# Baseline Accuracy
df_raw['wage'].value_counts(normalize = True)
```

```python
dummy_list = ['workclass', 'education', 'marital-status', 'occupation',
              'relationship', 'sex', 'native-country', 'wage']
df = pd.get_dummies(df_raw, columns= dummy_list, drop_first=True)
# df.head()
```

```python
plt.figure(figsize= (10, 50))
sns.heatmap(df.corr().sort_values('wage_ >50K', ascending = False)[['wage_ >50K']],
            vmin = -1,
            vmax = 1, 
            annot= True,
            cmap = 'viridis')
```

```python
X = df.drop(columns=['wage_ >50K'])
y = df['wage_ >50K']

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    stratify= y,
                                                    random_state = 42)

ss = StandardScaler()

X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

lr = LogisticRegressionCV(cv=10, max_iter=500)

lr.fit(X_train_scaled, y_train)
```

```python
print(lr.score(X_train_scaled, y_train))
lr.score(X_test_scaled, y_test)
```

```python
coef_df = pd.DataFrame({
    'value': lr.coef_[0],
    'variable': X.columns
})

coef_df_imp = coef_df[(coef_df['value'] >= 0.02) | (coef_df['value'] <= -0.02)]

lasso_features = [each for each in coef_df_imp['variable'].values]

df_lasso = df[lasso_features]
```

```python
df_lasso.shape

X = df_lasso
y = df['wage_ >50K']

X_train_lasso, X_test_lasso, y_train_lasso, y_test_lasso = train_test_split(X, 
                                                    y, 
                                                    stratify= y,
                                                    random_state = 42)

ss = StandardScaler()

X_train_lasso_scaled = ss.fit_transform(X_train)
X_test_lasso_scaled = ss.transform(X_test)

lr = LogisticRegressionCV(cv=10, max_iter=500)

lr.fit(X_train_scaled, y_train)
```

```python
lr.score(X_train_lasso_scaled, y_train)
```

```python
lr.score(X_test_scaled, y_test)
```

# kNN & Naive Bayes

# Boosting Models

### AdaBoost

```python
ada = AdaBoostClassifier(base_estimator=RandomForestClassifier(), random_state=42)
ada_params = {
   'n_estimators': [100, 159],
   'base_estimator__max_depth': [1, 2],
   'learning_rate': [1, 1.2]
}

gs = GridSearchCV(ada, param_grid=ada_params, cv=3)
gs.fit(X_train, y_train)

gs.best_params_

gs.predict(X_test)

gs.best_score_

gs.score(X_test, y_test)
```

### Gradient Boost

```python
gboost = GradientBoostingClassifier(random_state=42)
gboost_params = {
   'max_depth': [4, 5],
   'n_estimators': [50, 75, 100],
   'learning_rate': [.1, .12, .15]
}

gb_gs = GridSearchCV(gboost, param_grid=gboost_params, cv=3)
gb_gs.fit(X_train, y_train)

gb_gs.best_params_

gb_gs.best_score_

gb_gs.score(X_test, y_test)
```
